### 1 Dimensional CA

(Code modified from P. Charbonneau's book [Natural Complexity](https://press.princeton.edu/books/hardcover/9780691176840/natural-complexity))

One dimensional cellular automata are generated from simple rules dictating which elements of a one-dimensional array are 'on' or 'off'.  We can visualize the evolution of these arrays by stacking them and using black and white for their state.  For a given cell, its evolution is determined by its neighboring cells.  For example, if we decide on the rule
<blockquote>A cell turns black if either or both of its immediate neighbors is black</blockquote>
we get the following pattern

```python
def cell_1d_rule250(num_iter, width):
    """
    Returns a time series of a 1D cellular automata
    
    Args:
        * num_iter: <int> Number of steps in the cellular evolution
        * width: <int> Size of the 1D cell-space
        
    Returns:
        A plot of the time series evolution of the 1D cellular automata
    """
    image = np.zeros([num_iter, width], dtype='int') # Initialize array
    image[0,int(width/2)] = 1 # IC: Set the middle cell to 'on'
    
    for j in range(num_iter-1): # Each successive time-step
        for i in range(1, width-1): # Each cell in the space
            if image[j, i-1] == 1 or image[j, i+1] == 1:
                image[j+1, i] = 1 # Rule 250
        image[j,0] = image[j,width-2] # Enforce Periodicity
        image[j,width-1] = image[j,1] # Enforce Periodicity
        
    plt.imshow(image, cmap='binary')#, interpolation="nearest")
    plt.title("Rule 250")
    plt.show()
    
cell_1d_rule250(50,102)
```

A rather simple checkerboard pattern.
Now, suppose we implement a rule such as:
<blockquote>A cell turns black if one and only one of its immediate neighbors is black</blockquote>
Can we predict the outcome?

```python
def cell_1d_rule90(num_iter, width):
    """
    Returns a time series of a 1D cellular automata
    
    Args:
        * num_iter: <int> Number of steps in the cellular evolution
        * width: <int> Size of the 1D cell-space
        
    Returns:
        A plot of the time series evolution of the 1D cellular automata
    """
    image = np.zeros([num_iter, width], dtype='int') # Initialize array
    image[0,int(width/2)] = 1 # IC: Set the middle cell to 'on'
    
    for j in range(num_iter-1): # Each successive time-step
        for i in range(1, width-1): # Each cell in the space
            if int(abs(image[j, i-1] - image[j, i+1])): # Rule 90
                image[j+1, i] = 1
        image[j,0] = image[j,width-2] # Enforce Periodicity
        image[j,width-1] = image[j,1] # Enforce Periodicity
        
    plt.imshow(image, cmap='binary')#, interpolation="nearest")
    plt.title("Rule 90")
    plt.show()
    
cell_1d_rule90(512,1000)
```

We see that this simple rule yields a surprisingly complex structure (knows as the Sierpenski Triangle).
Finally, consider a slightly different rule with an asymmetry:
<blockquote>A cell turns black if: only one of itself and its immediate neighbors is black; OR only itself and its right neighbor is black</blockquote>


```python
def cell_1d_rule30(num_iter, width):
    """
    Returns a time series of a 1D cellular automata
    
    Args:
        * num_iter: <int> Number of steps in the cellular evolution
        * width: <int> Size of the 1D cell-space
        
    Returns:
        A plot of the time series evolution of the 1D cellular automata
    """
    image = np.zeros([num_iter, width], dtype='int') # Initialize array
    image[0,int(width/2)] = 1 # IC: Set the middle cell to 'on'
    
    for j in range(num_iter-1): # Each successive time-step
        for i in range(1, width-1): # Each cell in the space
            if image[j, i-1] + image[j, i] + image[j, i+1] == 1: # Rule 30
                image[j+1, i] = 1
            elif image[j, i-1] == 0 and image[j, i] + image[j, i+1] == 2:
                image[j+1, i] = 1
        image[j,0] = image[j,width-2] # Enforce Periodicity
        image[j,width-1] = image[j,1] # Enforce Periodicity
        
    plt.imshow(image, cmap='binary')#, interpolation="nearest")
    plt.title("Rule 30")
    plt.show()
    
cell_1d_rule30(200,410)
```

This rule generates a pattern that has both ordered and disordered characteristics.  In all of these examples, a simple change of the generating rules produced a dramatic change in the overall pattern.
An interesting observation made by Stephen Wolfram is that the generating rules can be characterized by an eight digit code:
Note that for the rules given above, the evolution of a given cell is completely determined by the state of itself and its two immediate neighbors.  That is, three bits ($2^3=8$ possibilities) are needed to determine the next step in an iteration.  Therefore, every possible rule can be given by specifying how each of these eight states determine the next step.  Therefore, there are $2^8=256$ possible rules.
Wolfram created a taxonomy for these automata by converting these rules from an 8-bit binary number to decimal:

| 111 | 110 | 101 | 100 | 011 | 010 | 001 | 000 |
|:-:|:-:|:-:|:-:|:-:|:-:|:-:|:-:|
|0|0|0|1|1|1|1|0|

$00011110 = 30$ in binary, so we call the last pattern *Rule 30*.

Furthermore, we note that these patterns are completely deterministic.  No matter how we start the automata, the same outcome must always follow. However, it is not necessarily the case that every automata will generate an arbitrary sequence during its evolution.  For example, from inspection of the overall pattern, is is clear that the rule 250 automata cannot generate a completely black sequence.

256 rules does not seem like an overwhelming number of possibilities. Let's see if we can learn anything by visualizing all of them:

### Visualization of CA variation
How to encode each rule in pythjon: represent each rule as an 8 bit number.  There should be 256 different ones
How to evolve an automata given an initial string.
This should be general enough that it takes an initial condition and can evolve it through a desired number of steps.
    We should be able to determine how much space to give the automata so we don't run into boundary issues:
    For the rightmost black cell, the pattern can only grow at a rate of 1 cell per evolution.
    Therefore, if the rightmost black is at index $k$ and we are only allowing for $t$ evolutionary steps, the space needs to be at least $k+t+2$ (for PBC).
    Similarly, if the leftmost black is at $l$, the space needs to be $l-t-2$
        **NOTE:** If we are going for a ML, we should have inputs and outputs of the same size.  Lets start out by defining the maximum size of the initial condition space (i.e. 10 cells), and then using this size with the number of evolutions to build a template for all future patterns.

```python
def build_automata(rule_decimal, num_steps, IC, fname, save=True):
    """
    Saves the time evolution of a given rule for a given initial condition.
    
    Params:
        * rule_decimal <int>: Wolfram CA code number
        * num_steps <int>: Number of time steps to evolve CA
        * IC <numpy 1d array>: Initial condition state, given as a 1d array of black (one) or white (zero).
        * fname <str>: filename to save the automata.
        
    Returns:
        * Nothing.  Saves the automata to filename as a numpy array.  Displays an image representation of the automata evolution.
    """
    # Convert decimal to binary and store as an array
    wolfram_num = np.array([int(i) for i in format(rule_decimal, '08b')])
    
    
    # Initialize time-space array for automata evolution
    width = len(IC) + 2*num_steps + 4
    image = np.zeros([num_steps+1, width], dtype='int') # Max 'speed' of a cell is 1 horizontal step per iteration (left and right x2).  Added a buffer (+4)
    image[0,:] = np.concatenate((np.zeros(num_steps+2), IC, np.zeros(num_steps+2)), axis=None) # Replace 'top' row with a buffered initial condition.

    
    # Set up rules for time evolution
    keys = np.array(['111','110','101','100','011','010','001','000'])
    rule = dict(zip(keys,wolfram_num)) # A dictionary with the parents as keys, and values as the Wolfram Number
    
    
    # Begin Evolution
    for j in range(1,num_steps+1): #each time step
        for i in range(1,np.shape(image)[1]-1): # Each cell in the time step
            precursor = image[j-1, i-1:i+2] # Look at the three cells that determine evolution
            image[j, i] = rule[''.join(map(str, precursor))] # Determine the new cell by referencing the dictionary
            image[j,0] = image[j,width-2] # Enforce Periodicity
            image[j,width-1] = image[j,1] # Enforce Periodicity
            
    if save:
        np.save(fname, image, allow_pickle=False)
    else:
        plt.imshow(image, cmap='binary')
        plt.title('Rule {}'.format(rule_decimal))
        plt.show()
        
```

## Supervised Neural Networks

Suppose that instead of looking at how a single cell evolves according to its rule, a random collection of cells is introduced and then evolved from a given rule.  Is it possible to train a neural network to guess the rule from a given image?

In order to apply a predictive neural network, we need to generate a set of training data.  However, it is a best practice to split up training data into two sets:  One set will be used to adjust the weights of the neural network, while the other will be used to *validate* the trained network on data to which the network has not been exposed.  This is done to prevent overfitting of the network, which is effectively the network 'memorizing' the training data, rather than learning from it.

Thus, we need to generate a large body of data to build a neural network.

To do so, we will first generate a large variety of such evolutions.  Since there are 256 such rules (though only 88 are fundamentally distinct), exploring the full space of a 50 cell starting condition will result in $2^{50} \times 256 \approx 3 \times 10^{17}$ images.  My computer lacks the storage for so much data, so lets only pull 300 randomly initialized evolutions for each rule.  This will yield 76800 images, which we will further subdivide into 225 evolutions per rule for training, 25 evolutions per rule for validation, and 50 evolutions per rule for testing.

### Generation of Training Data

```python
### CAUTION:  THIS WILL TAKE A VERY LONG TIME TO RUN!

# Specify *PATH* to folder where you want to save training data

IC_len = 50
path = './Data_IC_{}_step_{}/'.format(IC_len, evolution_steps)
path_trng = path+'/Training/'
path_valid = path+'/Validation/'
path_test = path+'/Testing/'

try:
    os.mkdir(path)
    os.mkdir(path_trng)
    os.mkdir(path_valid)
    os.mkdir(path_test)
except OSError as error:
    pass

for j in range(256): # Iterate over all rules
    i = 0
    path = path_trng
    while True:
        if i==225:
            path = path_valid
        elif i==250:
            path = path_test
        elif i==300:
            break
            
        IC = np.random.randint(2,size=IC_len)
        IC_name = ''.join(map(str, IC))
        fname = format(j, '03d') + '_' + IC_name + '.npy'
        
        if os.path.isfile(path_trng+fname) or os.path.isfile(path_valid+fname) or os.path.isfile(path_test+fname):
            continue
        else:
            build_automata(j, evolution_steps, IC, path+fname)
            i += 1
```

Now that we have generated the data, we need to be able to turn it into a form our network can handle.

```python
def convert_data(directory):
    """
    Converts a directory of data into a list of tuples, 
    each of which is a 1d array consisting of a flattened image coupled with
    a unit vector in a 256 dimensional space representing its CA rule.
    """
    num_samples = len([file for file in os.listdir(directory) if os.path.isfile(directory+file)])
    height, width = np.load(directory+os.listdir(directory)[0]).shape
    data=[]
    labels=[]
    for file in os.listdir(directory):
        data.append(np.load(directory+file)) # list.append is faster than np.append
        labels.append(int(file[:3]))
    data = np.array(data)
    labels = np.array(labels)
    return data, labels

# Long time to run.  Load from saved file preferred
    
trng_inputs, trng_labels = convert_data('./Data_IC_50_step_64/Training/')
valid_inputs, valid_labels = convert_data('./Data_IC_50_step_64/Validation/')
test_inputs, test_labels = convert_data('./Data_IC_50_step_64/Testing/')

path_processed = path+'Processed/'
data_dict = {'trng_inputs':trng_inputs, 'trng_labels':trng_labels, 'valid_inputs':valid_inputs, 'valid_labels':valid_labels, 'test_inputs':test_inputs, 'test_labels':test_labels}

try:
    os.mkdir(path_processed)
except OSError as error:
    pass

for fname in data_dict.keys():
    np.save(path_processed+fname, data_dict[fname], allow_pickle=False)

# Uncomment the below if you have already processed the ML data    

# trng_inputs = np.load(path_processed+'trng_inputs.npy')
# trng_labels = np.load(path_processed+'trng_labels.npy')
# valid_inputs = np.load(path_processed+'valid_inputs.npy')
# valid_labels = np.load(path_processed+'valid_labels.npy')
# test_inputs = np.load(path_processed+'test_inputs.npy')
# test_labels = np.load(path_processed+'test_labels.npy')
```

```python
# Now condition the labels using Keras' one-hot encoding

num_classes = 256
trng_labels_cat = keras.utils.to_categorical(trng_labels, num_classes)
valid_labels_cat = keras.utils.to_categorical(valid_labels, num_classes)
test_labels_cat = keras.utils.to_categorical(test_labels, num_classes)
```

Before we start building our neural network, lets build a useful plotter to show the performance of our network.  This code comes courtesy of [mjbhobe](https://github.com/mjbhobe/dl-tensorflow-keras).

```python
def show_plots(history, plot_title=None, fig_size=None):
    """ Useful function to view plot of loss values & accuracies across the various epochs
        Works with the history object returned by the train_model(...) call """
    assert type(history) is dict

    # NOTE: the history object should always have loss & acc (for training data), but MAY have
    # val_loss & val_acc for validation data
    loss_vals = history['loss']
    val_loss_vals = history['val_loss'] if 'val_loss' in history.keys() else None
    epochs = range(1, len(history['categorical_accuracy']) + 1)

    f, ax = plt.subplots(nrows=1, ncols=2, figsize=((16, 4) if fig_size is None else fig_size))

    # plot losses on ax[0]
    ax[0].plot(epochs, loss_vals, color='navy', marker='o', linestyle=' ', label='Training Loss')
    if val_loss_vals is not None:
        ax[0].plot(epochs, val_loss_vals, color='firebrick', marker='*', label='Validation Loss')
        ax[0].set_title('Training & Validation Loss')
        ax[0].legend(loc='best')
    else:
        ax[0].set_title('Training Loss')

    ax[0].set_xlabel('Epochs')
    ax[0].set_ylabel('Loss')
    ax[0].grid(True)

    # plot accuracies
    acc_vals = history['categorical_accuracy']
    val_acc_vals = history['val_categorical_accuracy']# if 'val_acc' in history.keys() else None

    ax[1].plot(epochs, acc_vals, color='navy', marker='o', ls=' ', label='Training Accuracy')
    if val_acc_vals is not None:
        ax[1].plot(epochs, val_acc_vals, color='firebrick', marker='*', label='Validation Accuracy')
        ax[1].set_title('Training & Validation Accuracy')
        ax[1].legend(loc='best')
    else:
        ax[1].set_title('Training Accuracy')

    ax[1].set_xlabel('Epochs')
    ax[1].set_ylabel('Accuracy')
    ax[1].grid(True)

    if plot_title is not None:
        plt.suptitle(plot_title)

    plt.show()
    plt.close()

    # delete locals from heap before exiting (to save some memory!)
    del loss_vals, epochs, acc_vals
    if val_loss_vals is not None:
        del val_loss_vals
    if val_acc_vals is not None:
        del val_acc_vals
```

### Building a neural network
Now that we have a collection of data, we can train a neural network.  Lets start with a single-layer dense neural network with 150 neurons.

```python
trng_inputs_dense = trng_inputs.reshape(len(trng_inputs), -1)
valid_inputs_dense = valid_inputs.reshape(len(valid_inputs), -1)

net_D1 = keras.models.Sequential() # Build a sequential network
net_D1.add(keras.layers.Dense(30, input_shape=(trng_inputs.shape[1]*trng_inputs.shape[2],), activation='relu')) # Add an input layer with 2 input neurons densely connected to 30 neurons and a Rectified Linear Unit activation function
net_D1.add(keras.layers.Dense(num_classes, activation='softmax')) # Add a final output layer

net_D1.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['categorical_accuracy']) # Compile the network with categorical cross-entropy cost function using an adaptive learning rate
print(net_D1.summary())
```

```python
results_D1 = net_D1.fit(trng_inputs_dense, trng_labels_cat, epochs=15, batch_size=64, validation_data=(valid_inputs_dense, valid_labels_cat))
i=1
while True:
    if i >=100:
        break
    elif not os.path.isfile('./Models/D1_'+format(i,'02d')):
        net_D1.save('./Models/D1_'+format(i,'02d'))
        break
    else:
        i+=1
```

```python
show_plots(results_D1.history)
```

This is a pretty poor performance for our NN.  We do not even seem to be converging towards anything better than 50% accuracy!  Lets try adding some layers.

```python
net_D2 = keras.models.Sequential() # Build a sequential network
net_D2.add(keras.layers.Dense(100, input_shape=(trng_inputs.shape[1]*trng_inputs.shape[2],), activation='relu')) # Add an input layer with 2 input neurons densely connected to 100 neurons and a Rectified Linear Unit activation function
net_D2.add(keras.layers.Dense(50, activation='relu'))
net_D2.add(keras.layers.Dense(num_classes, activation='softmax')) # Add a final output layer

net_D2.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['categorical_accuracy']) # Compile the network with categorical cross-entropy cost function using an adaptive learning rate
print(net_D2.summary())
```

```python
results_D2 = net_D2.fit(trng_inputs_dense, trng_labels_cat, epochs=15, batch_size=64, validation_data=(valid_inputs_dense, valid_labels_cat))
i=1
while True:
    if i >=100:
        break
    elif not os.path.isfile('./Models/D2_'+format(i,'02d')):
        net_D2.save('./Models/D2_'+format(i,'02d'))
        break
    else:
        i+=1
```

```python
show_plots(results_D2.history)
```

This seems like we are doing better, but we note the validation accuracy starts decreasing.   This is a sign of overfitting the data.  That is, we are training our neural network to memorize the training set, rather than deduce fundamental features of the training data.  Unfortunately, it seems like we cannot solve this problem with dense networks.  If a dense network has $M$ input neurons, $N$ neurons in the first hidden layer, ..., and $T$ neurons in the output layer, the network will have $M\times N \times \cdots \times T$ weights.  This makes adding neurons or layers computationally prohibitive.

However, if we apply some forethought to our problem, we realize that a given cell can only depend on its ancestors nearest neighbors.  There is no need for the network to try to draw patterns relating cells on opposite ends of the space!  In fact, if we were to try to solve this problem 'manually', a reasonable strategy would be to look at the three ancestors directly above it, as well as the two neigbors in its generation to see how they relate to each other.  This is the whole point behind a convolutional neural network!

```python
trng_inputs_conv = np.reshape(trng_inputs, (trng_inputs.shape[0], trng_inputs.shape[1], trng_inputs.shape[2], 1))
valid_inputs_conv = np.reshape(valid_inputs, (valid_inputs.shape[0], valid_inputs.shape[1], valid_inputs.shape[2], 1))
test_inputs_conv = np.reshape(test_inputs, (test_inputs.shape[0], test_inputs.shape[1], test_inputs.shape[2], 1))
```

```python
net_C1 = keras.models.Sequential() # Build a sequential network
net_C1.add(keras.layers.Conv2D(input_shape=(trng_inputs.shape[1],trng_inputs.shape[2],1), filters=8, kernel_size=[3,3], activation='relu', padding='same')) # Expect there to be 8 distinct patterns, with a 3x3 kernel
net_C1.add(keras.layers.MaxPooling2D(pool_size=(3,3)))
net_C1.add(keras.layers.Conv2D(filters=8, kernel_size=[3,3], activation='relu', padding='same'))
net_C1.add(keras.layers.MaxPooling2D(pool_size=(3,3)))
net_C1.add(keras.layers.Flatten())
net_C1.add(keras.layers.Dense(50, activation='relu'))
net_C1.add(keras.layers.Dense(num_classes, activation='softmax')) # Add a final output layer

net_C1.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['categorical_accuracy']) # Compile the network with categorical cross-entropy cost function using an adaptive learning rate
print(net_C1.summary())
```

```python
results_C1 = net_C1.fit(trng_inputs_conv, trng_labels_cat, epochs=15, batch_size=64, validation_data=(valid_inputs_conv, valid_labels_cat))
i=1
while True:
    if i >=100:
        break
    elif not os.path.isfile('./Models/C1_'+format(i,'02d')):
        net_C1.save('./Models/C1_'+format(i,'02d'))
        break
    else:
        i+=1
```

```python
show_plots(results_C1.history)
```

This looks pretty good! We see that the validation accuracy tracks well with the training accuracy at about $~96\%-97\%$.  Importantly, it does not decrease, suggesting that we have not reached the 'over-fitting' regime.  Lets use this model on our test data:

```python
test_loss, test_accuracy = net_C1.evaluate(test_inputs_conv, test_labels_cat, batch_size=64)
print('Test loss: %.4f accuracy: %.4f' % (test_loss, test_accuracy))
```

```python
predictions_probs=net_C1.predict_on_batch(test_inputs_conv[0:30])
predictions = np.argmax(predictions_probs, axis=1)
true_labels = np.argmax(test_labels_cat[0:30], axis=1)
print("Predictions: ", predictions)
print("True Labels: ", true_labels)
```

Pretty not bad for a relatively simple neural network!
